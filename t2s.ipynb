```python
import json

# Specify the file path
file_path = rf"C:\Users\Owner\anaconda3\envs\art10\Lib\site-packages\sign_language_translator\assets\pk-dictionary-mapping.json"


# Read the JSON file
with open(file_path, 'r') as f:
    data = json.load(f)

# Access the parsed data
words = data[0]['mapping']
```

```python
for i in words:
    i['token']['ur'] = [word.lower() for word in i['token']['ur']]
```

```python
def get_video_from_word(word):
    for i in words:
        if word.lower() in i['token']['ur']:
            return [i['label']]
    return None
```

```python
import random
def get_character (letter):

    video = get_video_from_word(letter)
    if video == None:
        video = random.choice(words)['label']

    return video
```

```python
def get_video_character_by_character (word):
    print (f"{word} not in words")
    word_list = []
    for i in word:
        word_list.extend(get_character(i))
    return word_list
```

```python
def get_video (word):
    video = get_video_from_word(word)
    if video == None:
        if len(word) > 5:
           video = random.choice(words)['label']
        else:
            video = get_video_character_by_character(word)
    return video
```

```python
def get_video_from_text (text):
    clean_text = re.sub(r'[^а-яА-Яa-zA-Z0-9\s]', '', text)

    # Split the text into words
    words = clean_text.split()

    # Convert all words to lowercase
    lowercase_words = [word.lower() for word in words]

    # Get the video for each word
    video_list = []
    for word in lowercase_words:
        # add all videos in get_video in the video list
        video_list.extend(get_video(word))
    
    return video_list
```

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips

# Create a list of video paths from the video names
video_paths = [rf"C:\Users\Owner\Documents\FMI\Hackaton\videos\{video}.mp4" for video in videos]
```

```python
import cv2

# Set the desired display size
k = 0.5
display_width = round (720 * k)
display_height = round (1280 * k)

print(video_paths)
flag = True
for video in video_paths:
    if flag == False:
        break
    cap = cv2.VideoCapture(video)
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Resize the frame
        resized_frame = cv2.resize(frame, (display_width, display_height))

        # Display the resized frame
        cv2.imshow('Video', resized_frame)

        if cv2.waitKey(25) & 0xFF == ord('q'):
            flag = False
            break
    cap.release()

cv2.destroyAllWindows()
```
